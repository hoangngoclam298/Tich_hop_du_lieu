```python
attr_data = ['title', 'price', 'address', 'area', 'date', 'description',
             'link_image', 'url_page', 'direct', 'floor', 'name_contact', 'phone_contact']
```

Train data: alomuabannhdat + ibatdongsan

```python
def preprocess_text(text):
    # Chuyển đổi các ký tự có dấu thành không dấu
    text = unidecode.unidecode(text)

    text = text.replace("-", " ").replace("/", " ").replace(".", " ")

    # Xóa các ký tự không phải chữ cái hoặc số
    text = re.sub(r'[^a-zA-Z0-9 ]', '', text)

    # Xóa khoảng trắng thừa
    text = re.sub(r'\s+', ' ', text).strip()

    return text
```

```python
data_train = {}
data_link = []
for attr in attr_data:
    tmp_data = []
    if(attr != 'link_image'):
        if(attr in homedy_data.columns):
            for x in homedy_data[attr]:
                if(type(x) == str):
                    if(x != ''):
                        tmp_data.append(preprocess_text(x))
                elif(type(x) == float):
                    if(not np.isnan(x)):
                        tmp_data.append(str(x))
        if(attr in i_data.columns):
            for x in i_data[attr]:
                if(type(x) == str):
                    if(x != ''):
                        tmp_data.append(preprocess_text(x))
                elif(type(x) == float):
                    if(not np.isnan(x)):
                        tmp_data.append(str(x))
    else:
        if(attr in homedy_data.columns):
            for x in homedy_data[attr]:
                for y in x:
                    data_link.append(preprocess_text(y))
        if(attr in i_data.columns):
            for x in i_data[attr]:
                for y in x:
                    data_link.append(preprocess_text(y))
    data_train[attr] = tmp_data
random.shuffle(data_link)
data_train['link_image'] = data_link[:14000]
```

```python
from sklearn.svm import SVC
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
import joblib

from sklearn.feature_extraction.text import CountVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score

for attr_train in attr_data:
    # Chuẩn bị dữ liệu và đặc trưng
    p_text = []
    n_text = []
    for attr in attr_data:
        if(attr == attr_train):
            p_text += data_train[attr]
        else:
            n_text += data_train[attr]

    ratio = len(p_text)/len(n_text)

    nn_text = []

    for i in range(len(n_text)):
        if(True):
            nn_text.append(n_text[i])

    print('Attr:', attr_train, '\nLen positive:', len(p_text), 'len negative:', len(nn_text))
    sentences = p_text + nn_text
    labels = [1]*len(p_text) + [0]*len(nn_text)

    vectorizer = TfidfVectorizer()
    X = vectorizer.fit_transform(sentences)

    # Chia dữ liệu thành tập huấn luyện và tập kiểm tra
    X_train, X_test, y_train, y_test = train_test_split(X, labels, test_size=0.2, random_state=42)

    # Xây dựng và huấn luyện mô hình SVM
    svm_model = SVC()
    svm_model.fit(X_train, y_train)

    # Dự đoán nhãn cho tập kiểm tra
    y_pred = svm_model.predict(X_test)

    # Đánh giá mô hình
    accuracy = accuracy_score(y_test, y_pred)
    print('SVM Accuracy:', accuracy)
    # Save model and vectorizer
    name_model = '/content/drive/MyDrive/2022_2/Data_Integration/Train_1_10/' + attr_train + '_model_SVM.pkl'
    name_vect = '/content/drive/MyDrive/2022_2/Data_Integration/Train_1_10/' + attr_train + '_vect.pkl'
    joblib.dump(svm_model, name_model)
    joblib.dump(vectorizer, name_vect)


    # Decision Tree
    dt_model = DecisionTreeClassifier()
    # Huấn luyện mô hình trên tập huấn luyện
    dt_model.fit(X_train, y_train)

    # Dự đoán nhãn cho tập kiểm tra
    y_pred = dt_model.predict(X_test)

    # Đánh giá độ chính xác của mô hình
    accuracy = accuracy_score(y_test, y_pred)
    print("DT Accuracy:", accuracy)
    name_model = '/content/drive/MyDrive/2022_2/Data_Integration/Train_1_10/' + attr_train + '_model_DT.pkl'
    joblib.dump(dt_model, name_model)

    # Naive Bayes
    vectorizer = CountVectorizer()
    X = vectorizer.fit_transform(sentences)

    # Chia dữ liệu thành tập huấn luyện và tập kiểm tra
    X_train, X_test, y_train, y_test = train_test_split(X, labels, test_size=0.2, random_state=42)

    # Khởi tạo và huấn luyện mô hình Naive Bayes
    nb_model = MultinomialNB()
    nb_model.fit(X_train, y_train)

    # Dự đoán nhãn cho dữ liệu kiểm tra
    y_pred = nb_model.predict(X_test)

    # Đánh giá độ chính xác của mô hình
    accuracy = accuracy_score(y_test, y_pred)
    print("NB Accuracy:", accuracy,'\n')
```
