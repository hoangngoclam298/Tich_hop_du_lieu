```python
def preprocess_text(text):
    # Chuyển đổi các ký tự có dấu thành không dấu
    text = unidecode.unidecode(text)

    text = text.replace("-", " ").replace("/", " ").replace(".", " ")

    # Xóa các ký tự không phải chữ cái hoặc số
    text = re.sub(r'[^a-zA-Z0-9 ]', '', text)

    # Xóa khoảng trắng thừa
    text = re.sub(r'\s+', ' ', text).strip()

    return text
```

```python
# list_data = [alo_data, homedy_data, i_data, mogi_data, nhadat_data]
attr_schema = ['title', 'price', 'address', 'area', 'date', 'description',
               'link_image', 'url_page', 'direct', 'floor', 'name_contact', 'phone_contact']
data_match = [alo_data, mogi_data, nhadat_data]
data_name = ['alo_data', 'mogi_data', 'nhadat_data']

data_text_match = {}
for i in range(len(data_match)):
    data = data_match[i]
    name = data_name[i]
    data_text = {}
    for attr in data.columns:
        tmp_data = []
        if(attr != 'link_image'):
            for x in data[attr]:
                if(type(x) == str):
                    if(x != ''):
                        tmp_data.append(preprocess_text(x))
                elif(type(x) == float):
                    if(not np.isnan(x)):
                        tmp_data.append(str(x))
        else:
            for x in data[attr]:
                for y in x:
                    tmp_data.append(preprocess_text(y))
                    break
        random.shuffle(tmp_data)
        if(len(tmp_data) > 1000):
            data_text[attr] = tmp_data[:1000]
        else:
            data_text[attr] = tmp_data
    data_text_match[name] = data_text

print(data_text_match)
```

```python
result_match = {}
for attr in attr_schema:
    name_svm_model = '/content/drive/MyDrive/2022_2/Data_Integration/Train_1_10/' + attr + '_model_SVM.pkl'
    name_dt_model = '/content/drive/MyDrive/2022_2/Data_Integration/Train_1_10/' + attr + '_model_DT.pkl'
    name_vect = '/content/drive/MyDrive/2022_2/Data_Integration/Train_1_10/' + attr + '_vect.pkl'

    svm_model = joblib.load(name_svm_model)
    dt_model = joblib.load(name_dt_model)
    vectorizer = joblib.load(name_vect)
    result_match[attr] = []
    for data in data_text_match:
        for attr_data in data_text_match[data]:
            list_text = data_text_match[data][attr_data]
            trans_text = vectorizer.transform(list_text)
            pred_svm = svm_model.predict(trans_text)
            pred_dt = dt_model.predict(trans_text)
            print('SVM', attr, attr_data, data, sum(pred_svm)/len(pred_svm))
            print('DT', attr, attr_data, data, sum(pred_dt)/len(pred_dt))
            result_match[attr].append(['SVM', data, attr_data, sum(pred_svm)/len(pred_svm)])
            result_match[attr].append(['DT', data, attr_data, sum(pred_dt)/len(pred_dt)])

print(result_match)
```

```python
import Levenshtein

def check_2text(str1, str2):
    # Tính khoảng cách Levenshtein giữa hai chuỗi
    distance = Levenshtein.distance(str1, str2)
    # Tính độ tương đồng
    similarity = 1 - (distance / max(len(str1), len(str2)))
    return similarity
```

```python
def cut_city(addr):
    if(type(addr) == float):
            return ''
    return addr.split(',')[-1]

def cut_district(addr):
    if(type(addr) == float):
            return ''
    if(len(addr.split(',')) < 2):
        return ''
    return addr.split(',')[-2]

data_final['city'] = data_final['address'].apply(cut_city)
data_city = data_final['city'].unique()
print(data_city)
len(data_city)

data_final['district'] = data_final['address'].apply(cut_district)
data_district = data_final['district'].unique()
print(data_district)
len(data_district)
```
